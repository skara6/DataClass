* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group

```python
purchase_data["Age"].drop_duplicates(keep=False,inplace=True) 
```

```python
bins = [0,10, 14, 19, 23, 27, 31, 35, 39, 43, 47 ]
age_grp=["0 to 10", "11 to 14", "15 to 19", "20 to 23", "24 to 27", "28 to 31", "32 to 35", "36 to 39","40 to 43", "44 to 47"]

# Slice the data and place it into bins


purchase_data["agegrp"]=pd.cut(purchase_data["Age"], bins, labels=age_grp)
purchase_data.head()

num_grp= purchase_data.groupby("agegrp")
tot_age=purchase_data["agegrp"].value_counts()
tot_age
comb_age=purchase_data["agegrp"].count()
comb_age
percent_age=tot_age/comb_age*100
percent_age
Demographics_df=pd.DataFrame(tot_age)
Demographics_df["Percentage"]=percent_age
Demographics_df["Percentage"]=Demographics_df["Percentage"].map("{:.2f}%".format)
Demographics_df
```

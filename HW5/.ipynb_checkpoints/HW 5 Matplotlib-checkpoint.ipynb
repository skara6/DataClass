```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load ="data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data=pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)
#ride_data.columns


# Combine the data into a single dataset
comb_city=pd.merge(ride_data,city_data, on="city")
comb_city.head(20)





# Display the data table for preview
```

```python
individual_city=comb_city['city'].value_counts()
group_city=comb_city.groupby(['city'])


#avg fare
avg_fare=group_city['fare'].mean()
avg_fare.min()

```

```python
#Total fares
total_fare=group_city['fare'].sum()
total_fare
```

```python
#TotalNumber of rides

tot_rides=group_city['ride_id'].count()
tot_rides.max()
```

```python
tot_drivers=group_city['driver_count'].sum()
tot_drivers
```

```python
city_type=comb_city['type'].value_counts()
city_type
```

```python
city_kind=group_city["type"].max()
city_kind
```

```python
city_df=pd.DataFrame({"Average Fare":avg_fare,
                     "Total Rides":tot_rides,
                     "Total Drivers":tot_drivers,
                     "Total Fare" : total_fare,
                   "City Type":city_kind })
city_df
```

```python
urban = city_df.loc[city_df['City Type'] == 'Urban']
total_fare_urban=urban["Total Fare"]
total_fare_urban
total_rides_urban = urban['Total Rides']
total_rides_urban
avg_fare_urban = urban['Average Fare']
avg_fare_urban
total_drivers_urban = urban['Total Drivers']
total_drivers_urban
```

```python
suburban = city_df.loc[city_df['City Type'] == 'Suburban']
total_fare_suburban=suburban["Total Fare"]
total_fare_suburban
total_rides_suburban = suburban['Total Rides']
total_rides_suburban
avg_fare_suburban = suburban['Average Fare']
avg_fare_suburban
total_drivers_suburban = suburban['Total Drivers']
total_drivers_suburban
```

```python
rural = city_df.loc[city_df['City Type'] == 'Rural']
total_fare_rural=rural["Total Fare"]
total_fare_rural
total_rides_rural = rural['Total Rides']
total_rides_rural
avg_fare_rural = rural['Average Fare']
avg_fare_rural
total_drivers_rural = rural['Total Drivers']
total_drivers_rural
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types 

plt.scatter(total_rides_urban, avg_fare_urban, marker="o", facecolors="LightSkyBlue", edgecolors="black",
            s=total_drivers_urban, alpha=0.75)


plt.scatter(total_rides_suburban, avg_fare_suburban, marker="o", facecolors="Lightcoral", edgecolors="black",
            s=total_drivers_suburban, alpha=0.75)

plt.scatter(total_rides_rural, avg_fare_rural, marker="o", facecolors="gold", edgecolors="black",
            s=total_drivers_rural, alpha=0.75)



#Urban = plt.scatter(total_rides_urban, avg_fare_urban, marker="o", facecolors="LightSkyBlue", edgecolors="black",
                    #s=total_drivers_urban, alpha=0.75)
#Suburban,= plt.scatter(total_rides_suburban, avg_fare_suburban, marker="o", facecolors="LightCoral", edgecolors="black",
                        #s=total_drivers_suburban, alpha=0.75)
#Rural,= plt.scatter(total_rides_rural, avg_fare_rural, marker="o", facecolors="gold", edgecolors="black",
                    #s=total_drivers_rural, alpha=0.75)


#plt.legend(handles=[Urban,Suburban,Rural],loc="upper right corner")


plt.ylim(15,50)
plt.xlim(0,50)
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.figtext(.95, .8, "NOTE: Circle size correlates with driver count")
plt.savefig("Images/ride_sharing.png")
plt.show()

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

## Total Fares by City Type

```python
 total_fares_bytype=city_df["Total Fare"].sum()
total_fares_bytype
```

```python
#Urban
percent_fares_urban=total_fare_urban.sum()/total_fares_bytype*100
percent_fares_urban

#Suburban
percent_fares_suburban=total_fare_suburban.sum()/total_fares_bytype*100
percent_fares_suburban

#Rural
percent_fares_rural=total_fare_rural.sum()/total_fares_bytype*100
percent_fares_rural

```

```python
# Show Figure

types=["Urban","Suburban","Rural"]
fares=[percent_fares_urban,percent_fares_suburban,percent_fares_rural]
colors=["LightCoral","LightSkyBlue","gold"]
explode=[0.1,0,0]

plt.pie(fares, explode=explode, labels=types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=250)
#plt.axis("equal")
plt.savefig("Images/total_fares")
plt.show()
```
